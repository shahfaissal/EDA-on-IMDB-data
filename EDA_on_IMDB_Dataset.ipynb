```python
#Required columns
columns_to_keep = ['budget','genres','imdb_id','original_title','original_language','revenue','cast','crew']
```

```python
df = df[columns_to_keep]
```

```python
#function to convert string to list
def string_to_list(str):
  return eval(str)
```

```python
#applying function to convert string to list
df.loc[~df['genres'].isna(),'genres'] = df.loc[~df['genres'].isna(),'genres'].apply(string_to_list)
df.loc[~df['cast'].isna(),'cast'] = df.loc[~df['cast'].isna(),'cast'].apply(string_to_list)
df.loc[~df['crew'].isna(),'crew'] = df.loc[~df['crew'].isna(),'crew'].apply(string_to_list)
```

```python
#creating a copy of dataset
df_new = df.copy()
```

# Here we are changing the data in budget column as budget of a movie cannot be below 1000, we are replacing it with the median value for the budget values and also for revenue column

```python
#Replacing extremely low values of budget and revenue with median values
df_new.loc[df_new['budget']<1000,'budget'] = df_new.loc[df_new['budget']>=1000,'budget'].median()
df_new.loc[df_new['revenue']<1000,'revenue'] = df_new.loc[df_new['revenue']>=1000,'revenue'].median()
```

```python
df_new.describe()
```

# profit can be calculated as the difference between revenue and budget

```python
#creating profit column for movies 
df_new['profit']=df_new['revenue'] - df_new['budget']
```

```python
df_new[['profit']]
```

1.Which movie made the highest profit? Who were its producer and director? Identify the actors in that film

```python
#movie with highest profit
highest_profit_id=df_new['profit'].argmax()
highest_profit = df_new['profit'].max()
print(f"The movie with highest profit is {df_new['original_title'][highest_profit_id]} with a profit of {highest_profit}")
```

```python
max_profit_movie_df = df_new[df_new['profit']==highest_profit]
```

```python
max_profit_movie_df
```

```python
#creating crew list
crew_list = max_profit_movie_df.iloc[0]['crew']
```

```python
#directors and producers list
directors = []
producers = []
for i in crew_list:
  if i['job']=='Producer':
    producers.append(i['name'])
  if i['job']=='Director':
    directors.append(i['name'])
```

```python
directors,producers
```

# Actors for highest grossing movie

```python
cast_list = max_profit_movie_df.iloc[0]['cast']
```

```python
#actor list
actors = []
for i in cast_list:
  actors.append(i['name'])
```
